```python
# Definimos nuestra funci√≥n para desanidar los valores de la columna genres
def buscar_name(obj): 
    if isinstance(obj, str) and '{' in obj:
        L=[]
        for i in ast.literal_eval(obj):
            L.append(i['name'])
        return L
```

```python
# Aplicamos nuestra funcion a la columna genres
df['genres'] = df['genres'].apply(buscar_name)
```

```python
df['genres'].head()
```

```python
# Obtenemos los valores de genres en una cadena de texto
df['genres'] = df['genres'].apply(lambda x: ', '.join(map(str, x)) if x is not None else None)
```

```python
#Corroboramos
df['genres'].head()
```

```python
# Definimos una nueva columna como g√©nero principal, teniendo en cuenta el primer valor.
df['genero_principal'] = df['genres'].str.split(',').str[0]
```

```python
# Corroboramos
df['genero_principal'][0]
```

```python
# Corroboramos valores nulos
df.isnull().sum()
```

```python
# Eliminamos valores nulos para las columnas genero pricipal y title.
df.dropna(subset=['genero_principal'], inplace=True)
```

```python
df.dropna(subset=['title'], inplace=True)
```

```python
df.isnull().sum()
```

```python
# Eliminamos los valores con formato jpg que contiene esta columna (se especifica tambien en el archivo PI1_ETL.ipynb),
# para poder convertirla en formato float
filtro = df['budget'].str.contains('.jpg')
```

```python
df = df[~filtro]
```

```python
df['budget'] = df['budget'].astype('float64')
```

```python
# Convertimos a formato numerico los datos de la columna id 
df['id'] = pd.to_numeric(df['id'], errors='coerce')
```

```python
# Convertimos a formato datetime los datos de la columna release_date
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
# Creamos una nueva columna release_year, a√±o de lanzamiento 
df['release_year'] = df['release_date'].dt.year
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Eliminamos datos nulos
df.dropna(subset=['release_year'], inplace=True)
```

```python
# Observamos  las correlaciones existentes entre algunas de las columnas de nuestro dataframe presupuesto, recaudaci√≥n y cantidad de votos
corr = df.corr()

sns.heatmap(corr, cmap='coolwarm')

plt.show()
```

Podemos observar que existe una muy buena correlaci√≥n entre los valores de las columnas budget y revenue, y budget y vote_count es una correlacion positiva. Esta correlaci√≥n tiene sentido, ya aquellas pel√≠culas a las cuales se les asigna un mayor presupuesto, son aquellas que tienen un mayor impacto, m√°s visualizaciones y por ende son aquellas que generan mayor recaudaci√≥n.

Podemos observar que los g√©neros principales que tienen mayor cantidad de pel√≠culas son Drama y Comedia.

```python
# Eliminamos valores nulos existentes en la columna vote_count
df.dropna(subset=['vote_count'], inplace=True)
```

```python
# Definimos un top 10 de las pel√≠culas con mayor cantidad de votos, eliminando aquellos valores duplicados.
top_diez = df.sort_values(by='vote_count', ascending=False).drop_duplicates('id')[:10]
```

```python
top_diez
```

```python
# Definimos una lista con nuestro top 10
movies_id = top_diez['id'].tolist()
```

```python
#Filtramos nuestro dataframe
df_filtrado = df[df['id'].isin(movies_id)]
```

```python
#Visualizamos la distribuci√≥n del n√∫mero de votaciones de las 10 peliculas mas votadas
df_filtrado['vote_count'].hist(bins=50)
plt.xlabel('N√∫mero de votaciones')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n del n√∫mero de votaciones de las 10 pel√≠culas m√°s votadas')
plt.show()
```

Podemos observar que en la distrubuci√≥n de frecuencias respecto a la cantidad de votaciones, se observa una mayor frecuencia en cantidades de votaciones menores. Esto podr√≠a deberse a que generalmente se producen una gran cantidad de pel√≠culas, que propablemente tengan un presupuesto bajo asignado, lo cual no genera mucho impacto, por ende tampoco gran cantidad de votaciones.

```python
# Importamos la librer√≠a wordcloud para poder visualizar la nube de palabras
from wordcloud import WordCloud
```

```python
#Ordenamos nuestro dataframe
df_top_diez = df.sort_values(by='vote_count', ascending=False).head(10)
```

```python
title_count = {}
for index, row in df_top_diez.iterrows():
    title_count[row['title']] = row['vote_count']
```

```python
wordcloud = WordCloud(background_color='white').generate_from_frequencies(title_count)
```

```python
# Visualizamos la nube de palabras
plt.figure(figsize=(12,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Podemos observar cu√°les son las pel√≠culas con mayor cantidad de votos, de una manera muy visual.
