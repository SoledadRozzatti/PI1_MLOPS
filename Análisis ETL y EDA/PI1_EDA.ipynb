```python
# Definimos nuestra función para desanidar los valores de la columna genres.

def buscar_name(obj):
    """
    Esta función devuelve una lista con los valores de la clave "name" de un objeto si es una cadena de texto
    y contiene caracteres "{". Si el objeto no cumple estas condiciones, retorna una lista vacía.

    Argumento:
        obj (str): El objeto a evaluar.

    Returns:
        list: Una lista con los valores de la clave "name" del objeto, o una lista vacía.

    Ejemplo:
        >>> buscar_name('{"name": "Juan"}')
        ['Juan']
    """
    if isinstance(obj, str) and "{" in obj:
        L = []
        for i in ast.literal_eval(obj):
            L.append(i["name"])
        return L
    else:
        return []
```

```python
# Aplicamos nuestra funcion a la columna genres.
df["genres"] = df["genres"].apply(buscar_name)
```

```python
# Importamos la librería wordcloud para poder visualizar la nube de palabras.
from wordcloud import WordCloud
```

```python
# Ordenamos nuestro dataframe.
df_top_diez = df.sort_values(by="vote_count", ascending=False).head(10)
```

```python
# Cálculo del recuento de votos por título.

''' Este bloque de código recorre cada fila del dataframe df_top_diez y crea un diccionario title_count
    donde la clave es el título de la película y el valor es el recuento de votos. El recuento de votos se 
    obtiene de la columna "vote_count" de cada fila del dataframe.

'''
title_count = {}
for index, row in df_top_diez.iterrows():
    title_count[row["title"]] = row["vote_count"]
```

```python
# Creación de un WordCloud a partir de las frecuencias de los títulos.
# Se crea un WordCloud a partir de las frecuencias de los títulos almacenadas en el diccionario title_count.
# El WordCloud representa visualmente las palabras más frecuentes en los títulos de películas, donde el tamaño de cada palabra está determinado por su frecuencia.
wordcloud = WordCloud(background_color='white').generate_from_frequencies(title_count)
```

```python
# Visualizamos la nube de palabras.
plt.figure(figsize=(12,8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

Podemos observar cuáles son las películas con mayor cantidad de votos, de una manera muy visual. Aquellas que aparecen resaltadas en un mayor tamaño, son aquellas que han recibido mayor cantidad de votos respectivamente.
