3- Desanidar los campos que contienen datos anidados, o bien buscar la manera de acceder a ellos sin desanidarlos. Con el fin de unirlos a un nuevo dataset para ser consumidos en consultas desde la API.  

```python
#Definimos funciones que van a extraer el dato name anidado
```

```python
def buscar_name(obj):
    """
    Esta función devuelve una lista con los valores de la clave "name" de un objeto si es una cadena de texto
    y contiene caracteres "{". Si el objeto no cumple estas condiciones, retorna una lista vacía.

    Argumento:
        obj (str): El objeto a evaluar.

    Returns:
        list: Una lista con los valores de la clave "name" del objeto, o una lista vacía.

    Ejemplo:
        >>> buscar_name('{"name": "Juan"}')
        ['Juan']
    """
    if isinstance(obj, str) and "{" in obj:
        L = []
        for i in ast.literal_eval(obj):
            L.append(i["name"])
        return L
    else:
        return []
```

```python
def buscar_name2(obj):
    """
    Esta función devuelve el valor de la clave "name" de un objeto si es una cadena de texto y contiene el carácter "{".
    Si el objeto no cumple estas condiciones o la clave "name" no está presente, retorna None.

    Argumento:
        obj (str): El objeto a evaluar.

    Returns:
        str or None: El valor de la clave "name" del objeto, o None si no cumple las condiciones o la clave "name" no está presente.

    Ejamplo:
        >>> buscar_name2('{"name": "Pedro"}')
        'Pedro'
    """
    if isinstance(obj, str) and "{" in obj:
        dic = ast.literal_eval(obj)
        return dic["name"]
```

```python
# Aplicamos las funciones definidas a las columnas anidadas
df["genres"] = df["genres"].apply(buscar_name)
df["belongs_to_collection"] = df["belongs_to_collection"].apply(buscar_name2)
df["production_companies"]  = df["production_companies"].apply(buscar_name)
df["production_countries"]  = df["production_countries"].apply(buscar_name)
df["spoken_languages"] = df["spoken_languages"].apply(buscar_name)
```

```python
# Convertimos los valores que obtuvimos de formato lista a formato no lista.
df["genres"] = df["genres"].apply(lambda x: ", ".join(map(str, x)) if x is not None else None)
df["production_companies"] = df["production_companies"].apply(lambda x: ", ".join(map(str, x)) if x is not None else None)
df["production_countries"] = df["production_countries"].apply(lambda x: ", ".join(map(str, x)) if x is not None else None)
df["spoken_languages"] = df["spoken_languages"].apply(lambda x: ", ".join(map(str, x)) if x is not None else None)
```

4- Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

```python
# Primero debemos transformar la columna budget a formato numerico.
# df['budget'] = pd.to_numeric(df['budget']) de esta forma lanza error porque hay registros con formato jpj.
```

```python
# Aplicamos un filtro para identificar los registros en formato jpg.
filtro = df["budget"].str.contains(".jpg")
```

```python
# Definimos nuevamente nuestro dataframe aplicando el filtro para que ya no contenga los registros en formato jpg.
df = df[~filtro]
```

```python
# Convertimos a formato float nuestra columna budget.
df["budget"] = df["budget"].astype("float64")
```

```python
# Corroboramos
df.info()
```

```python
# Creamos la nueva columna return solicitada.
# Definimos una función que calcula el retorno de inversión, teniendo en cuenta si existen nulos en numerador o denominador, y si el denominador es cero.

def calcular_retorno(row):
    """
    Esta función calcula el retorno dividiendo la columna 'revenue' entre la columna 'budget' de una fila.

    Argumento:
        row (Series): Una fila del DataFrame con las columnas 'revenue' y 'budget'.

    Returns:
        float: El valor del retorno calculado. Si alguna de las columnas es nula o el presupuesto es cero, retorna 0.

    Ejemplo:
        >>> calcular_retorno(pd.Series({'revenue': 1000000, 'budget': 500000}))
        2.0
    """
    revenue = row["revenue"]
    budget = row["budget"]
    if pd.isnull(revenue) or pd.isnull(budget) or budget == 0:
        return 0
    return revenue / budget

df["retorno"] = df.apply(calcular_retorno, axis=1)
```

7- Crear seis funciones para los endpoints que se consumirán en la API.

```python
df = pd.read_csv("Datasets\\dataset_ETL.csv")
df["release_date"] = pd.to_datetime(df["release_date"], format="%Y-%m-%d")
df["month_release"] = df["release_date"].dt.strftime("%B")
df["day_of_week_release"] = df["release_date"].dt.strftime("%A")


def peliculas_mes(mes: str):
    '''
    Esta función reciba como argumento el mes en español y la función retorna la cantidad de películas que se estrenaron ese mes
    (nombre del mes en español, en str, ejemplo "marzo") históricamente.

    Argumento:
        mes (str): El nombre del mes en español.

    Returns:
        dict: Un diccionario con el nombre del mes y la cantidad de películas estrenadas en ese mes.

    Ejemplo:
        >>> peliculas_mes("marzo")
        {"mes": "marzo", "cantidad": 10}
    '''
    meses = {
        "enero": "January",
        "febrero": "February",
        "marzo": "March",
        "abril": "April",
        "mayo": "May",
        "junio": "June",
        "julio": "July",
        "agosto": "August",
        "septiembre": "September",
        "octubre": "October",
        "noviembre": "November",
        "diciembre": "December"
    }
    mes_ingles = meses.get(mes.lower())
    df_mes = df[df["month_release"] == mes_ingles]
    cantidad = len(df_mes)
    return {"mes": mes, "cantidad": cantidad}



def peliculas_dia(dia: str):
    '''
    Esta función recibe como argumento el día de la semana y la función retorna la cantidad de películas que se estrenaron
    ese día (nombre del día en español, en str, ejemplo "lunes") históricamente.

    Argumento:
        dia (str): El nombre del día de la semana en español.

    Returns:
        dict: Un diccionario con el nombre del día y la cantidad de películas estrenadas en ese día.

    Ejemplo:
        >>> peliculas_dia("lunes")
        {"dia": "lunes", "cantidad": 5}
    '''
    dias_semana = {
        "lunes": "Monday",
        "martes": "Tuesday",
        "miercoles": "Wednesday",
        "jueves": "Thursday",
        "viernes": "Friday",
        "sabado": "Saturday",
        "domingo": "Sunday"
    }
    dia_ing = dias_semana[dia.lower()]
    
    df_dia = df[df["day_of_week_release"] == dia_ing]
    cantidad = len(df_dia)
    return {"dia": dia, "cantidad": cantidad}



def franquicia(franquicia: str):
    '''
    Esta función recibe como argumento la franquicia y retorna la cantidad de películas, la ganancia total
    y la ganancia promedio para dicha franquicia.

    Argumento:
        franquicia (str): El nombre de la franquicia.

    Returns:
        dict: Un diccionario con el nombre de la franquicia, la cantidad de películas, la ganancia total y la
        ganancia promedio.

    Ejemplo:
        >>> franquicia("Marvel Cinematic Universe")
        {"franquicia": "Marvel Cinematic Universe", "cantidad": 10, "ganancia_total": 2000000000, "ganancia_promedio": 200000000}
    '''
    df["belongs_to_collection"] = df["belongs_to_collection"].fillna("")
    df_franquicia = df[df["belongs_to_collection"].str.contains(franquicia, na=False)]
    cantidad = df_franquicia.shape[0]
    ganancia_total = df_franquicia["revenue"].sum()
    ganancia_promedio = df_franquicia["revenue"].mean()
    return {"franquicia": franquicia, "cantidad": cantidad, "ganancia_total": round(ganancia_total, 2), "ganancia_promedio": round(ganancia_promedio, 2)}



def peliculas_pais(pais: str):
    '''
    Esta función recibe como argumento el nombre del país y retorna la cantidad de películas producidas en el mismo.

    Argumento:
        pais (str): El nombre del país.

    Returns:
        dict: Un diccionario con el nombre del país y la cantidad de películas producidas en él.

    Ejemplo:
        >>> peliculas_pais("Estados Unidos")
        {"pais": "Estados Unidos", "cantidad": 1000}
    '''
    cantidad = len(df[df["production_countries"] == pais])
    return {"pais": pais, "cantidad": cantidad}



def productoras(productora: str):
    '''
    Esta función recibe como argumento el nombre de la productora y retorna la ganancia total y la cantidad de películas que produjeron.

    Argumento:
        productora (str): El nombre de la productora.

    Returns:
        dict: Un diccionario con el nombre de la productora, la ganancia total y la cantidad de películas producidas.

    Ejemplo:
        >>> productoras("Warner Bros")
        {"productora": "Warner Bros", "ganancia_total": 1000000000, "cantidad": 10}
    '''
    df_productora = df[df["production_companies"] == productora]
    ganancia_total = df_productora["revenue"].sum()
    cantidad = len(df_productora)
    return {"productora": productora, "ganancia_total": ganancia_total, "cantidad": cantidad}



def retorno(pelicula: str):
    '''
    Esta función recibe como argumento el nombre de la película y retorna la inversión, la ganancia, el retorno y el año en el que se lanzó.

    Argumento:
        pelicula (str): El título de la película.

    Returns:
        dict: Un diccionario con el nombre de la película, la inversión, la ganancia, el retorno y el año de lanzamiento.

    Ejemplo:
        >>> retorno("Avengers: Endgame")
        {"pelicula": "Avengers: Endgame", "inversion": 300000000, "ganancia": 2797800564, "retorno": 9.33, "anio": 2019}
    '''
    pelicula_df = df.loc[df["title"] == pelicula]
    inversion = pelicula_df["budget"].values[0]
    ganancia = pelicula_df["revenue"].values[0]
    retorno = pelicula_df["retorno"].values[0]
    anio = int(pelicula_df["release_date"].dt.year.values[0])
    return {"pelicula": pelicula, "inversion": inversion, "ganacia": ganancia, "retorno": retorno, "anio": anio}


```
